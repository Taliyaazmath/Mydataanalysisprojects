What is the Mean Value of each column against each "Weather Condition"?

```python
weather_means = {}

# Get the unique weather conditions
weather_conditions = data['Weather Condtion'].unique()

# Loop through each weather condition
for weather in weather_conditions:
    # Filter the data by the current weather condition
    weather_data = data[data['Weather Condtion'] == weather]

    # Calculate the mean of each numeric column
    weather_means[weather] = weather_data.mean(numeric_only=True)
#print(weather_means)
# Convert the dictionary to a DataFrame
weather_means_df = pd.DataFrame(weather_means).transpose()

# Print the DataFrame
print(weather_means_df.to_string())
```
